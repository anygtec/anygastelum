```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from sklearn import linear_model
import pandas as pd
```

```python
df = pd.read_csv('datasets/reg_data.csv')
df
```

```python
df.dropna(inplace= True)
df
```

```python
y= pd.DataFrame(df['Salary'])
x =pd.DataFrame(df['Age'])
lm= linear_model.LinearRegression()
model_lm=lm.fit(x,y)
model_lm
```

```python
print('The y intercetpt, B0 is =' , model_lm.intercept_)
print('the coef., b1 is =' , model_lm.coef_)
print('theR^2, is =', model_lm.score(x,y))
```

```python
#Salary= 0.380484541757693+48.09321584* Age
```

```python
modelA = sm.OLS.from_formula('Salary ~ Age', data = df)
resultA = modelA.fit()
resultA.summary()
```

1. What is the salary of a person with age 33?

```python
print('the salary of a person with age 33 is:' ,349.4458+48.0932*33)
```

```python
resiA = resultA.resid
fitValues= resultA.fittedvalues

resiA
fitValues
```

```python
plt.figure(figsize=(10,8))

ax= sns.regplot(x='Age',
               y='Salary',
               data=df,
               color = 'green',
                marker= '+',
                scatter_kws={'s':200},
              )
ax.set(xlabel = 'Age',
       ylabel = 'Salary')
ax.text(45,1500,'Salary = 349.4458+48.0932*Age',
        horizontalalignment = 'center', size= 'medium',
        color= 'black', weight= 'semibold')

ax.set_title('Regression plot of age vs Salary')
plt.show()
```
