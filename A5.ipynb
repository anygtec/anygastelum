```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('antropometria-dataset-1.csv')
df.head(2)
```

```python
sns.boxplot(df.cintura)
```

```python
pvedad = pd.pivot_table(df,
                       index = 'sexo',
                       aggfunc = 'median',
                       values = 'edad')
pvedad
```

```python
pvcintura = pd.pivot_table(df,
                       index = 'sexo',
                       aggfunc = 'median',
                       values = 'cintura')
pvcintura
```

```python
pvcintura = pd.pivot_table(df,
                       index = 'sexo',
                       aggfunc = 'min',
                       values = 'cintura')
pvcintura
```

```python
pvropa = pd.pivot_table(df,
                       index = ['sexo','desc_ent'],
                       aggfunc = 'sum',
                       values = 'ropa',
                       sort=)
pvropa
```

```python
pvropa.plot(kind='bar')
```
